```python
print("Code started")
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
print("code ended")
```

```python
print("Code started")
DATASET_PATH1= r'C:\Users\Lenovo\Desktop\sign_language\SignAlphaSet\SignAlphaSet\SignAlphaSet'
classes=sorted(os.listdir(DATASET_PATH))
print("Classes found: ",classes)
print("code ended")
```

```python
print("code started")
image_size=64
X_images=[]
y_labels=[]
for label in classes:
    folder_path=os.path.join(DATASET_PATH,label)
    for img_file in os.listdir(folder_path):
        img_path=os.path.join(folder_path,img_file)
        img=cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img=cv2.resize(img,(image_size,image_size))
            X_images.append(img)
            y_labels.append(classes.index(label))
print("code ended")            
       
```

```python
print("Total samples:", len(X_images))
print("First image shape:", X_images[0].shape)
print("First label:", y_labels[0])
```

```python
plt.imshow(X_images[0], cmap='gray')
plt.title(f"Label: {classes[y_labels[0]]}")
plt.axis('off')
plt.show()
```

```python
print("code started")
print("Converting data to NumPy arrays and one-hot encoding labels")
X = np.array(X_images).reshape(-1, image_size, image_size, 1).astype("float32") / 255.0
Y = to_categorical(y_labels, num_classes=len(classes))
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
print("Unique integer labels found in y:", set(y_labels))
print("code ended")
```

```python
print("code started")
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print("code ended")
```

```python
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_train[i].reshape(image_size,image_size),cmap='gray')
    plt.title(classes[np.argmax(Y_train[i])])
    plt.axis('off')
plt.tight_layout
plt.show
```

```python
#to see one image per class
plt.figure(figsize=(15, 8))
shown = set()
count = 0
for i in range(len(X)):
    label_index = y_labels[i]
    if label_index not in shown:
        shown.add(label_index)
        plt.subplot(4, 7, count+1)
        plt.imshow(X[i].reshape(image_size, image_size), cmap='gray')
        plt.title(classes[label_index])
        plt.axis('off')
        count += 1
    if count == 26:  # one for each class
        break
plt.tight_layout()
plt.show()
```

```python
print("code started")
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
print("code ended")
```

```python
print("code started")
from tensorflow.keras.layers import BatchNormalization, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(image_size, image_size, 1)),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.3),  # added here

    Conv2D(64, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.3),  # added here

    Conv2D(128, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.3),  # added here

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),  # already here  good
    Dense(len(classes), activation='softmax')
])
print("code ended")
```

```python
#compile
from tensorflow.keras.optimizers import Adam

print("code started")
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
print("code ended")
```

```python
#training
print("code started")
from tensorflow.keras.callbacks import EarlyStopping
early_stop =EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)
#augmentation
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    horizontal_flip=False,  # flipping might confuse similar signs
    validation_split=0.2
)

train_generator = datagen.flow(X, Y, batch_size=64, subset='training')
val_generator = datagen.flow(X, Y, batch_size=64, subset='validation', shuffle=False)
#using learning rate scheduler 
from tensorflow.keras.callbacks import ReduceLROnPlateau
lr_schedule = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2)
#actual training happens
history = model.fit(
    train_generator,
    epochs=20,
    validation_data=val_generator,
    callbacks=[early_stop, lr_schedule]
)
print("code ended")
```

```python
#evaluating the model
loss, accuracy = model.evaluate(X, Y)
print(f"Training Accuracy: {accuracy*100:.2f}%")
loss, accuracy = model.evaluate(val_generator)
print(f"Validation Accuracy: {accuracy*100:.2f}%")
```

```python
print("Starting Sign Language Recognition via Webcam...")

import cv2
import numpy as np
from tensorflow.keras.models import load_model
import pickle

# Load the trained model
model = load_model('sign_language_model.keras')

# Load the class names used during training
with open('class_labels.pkl', 'rb') as f:
    classes = pickle.load(f)  # Must match training order

# Define image size (must match training input size)
image_size = 64

# Start webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print(" Error: Could not access the webcam.")
    exit()

print(" Webcam started. Show a sign. Press 'q' to exit.")

while True:
    ret, frame = cap.read()
    if not ret:
        print(" Failed to capture frame.")
        break

    # Preprocess the frame
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(gray, (image_size, image_size))
    normalized = resized / 255.0
    input_data = normalized.reshape(1, image_size, image_size, 1)

    # Predict
    prediction = model.predict(input_data, verbose=0)
    predicted_index = np.argmax(prediction)
    predicted_label = classes[predicted_index]
    confidence = prediction[0][predicted_index]

    # Display result
    cv2.putText(frame, f"{predicted_label} ({confidence*100:.1f}%)",
                (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 2)
    cv2.imshow("Sign Language Detector", frame)

    # Break if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
print(" Detection ended.")
```
